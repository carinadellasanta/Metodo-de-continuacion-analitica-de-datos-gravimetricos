## Campo Anómalo verdadero VS Campo Anómalo por Continuación Analítica

**El objetivo es testear el Método de continuación analítica comparando sus resultados con los reales a los que debería aproximar.**

Para esto vamos a trabajar con dos esferas de radio 200 m, con contraste de densidad de 500 kg.m^3, con centros en las coordenadas x = +/-240 m, y = 0 m y z = -400 m.

El trabajo va a estar dividido en tres partes:

1. Cálculo del campo anómalo verdadero para 3 alturas z distintas: $z1=0m$, $z2=100m$ y $z3=-20m$
2. Cálculo del campo anómalo por continuación analítica ascendente para $Dz=100m$ y su comparación con el campo para $z=100m$ 
3. Cálculo del campo anómalo por continuación analítica ascendente para $Dz=-20m$ y su comparación con el campo para $z=-20m$ 

```python
#%pip install nbimporter
```

```python
import numpy as np                 # para calcular, numerical python.
import matplotlib.pyplot as plt    # para graficar, matplotlib.
import nbimporter                  # para importar otra note
import funciones_utiles
```

```python
rad  = 200.                    # radio de las esferas                       [m]
Dden = 500.                    # contraste de densidad                      [kg/m3]
x1   = 240.                    # centro de coordenada x de la esfera 1      [m]
x2   = -240.                   # centro de coordenada x de la esfera 2      [m]
y1   = 0.                      # centro de coordenada y de la esfera 1      [m]
y2   = 0.                      # centro de coordenada y de la esfera 2      [m]
z1   = -400.                   # centro de coordenada z de la esfera 1      [m]
z2   = -400.                   # centro de coordenada z de la esfera 2      [m]

x    = np.arange(-1100,1100,20)  # coordenadas x donde calculamos todas las anomalías [m]
y    = np.arange(-1100,1100,20)  # coordenadas y donde calculamos todas las anomalías [m]
```

## Cálculo de la anomalía vertical real producida por dos esferas de radio 200 m

**Calculo la anomalía vertical producida por dos esferas de radio 200 m para z=0**

Llamando a la función anom_esf definida en la note funciones_utiles

```python
zA   = 0.                      # coordenada z donde calculamos la anomalía  [m]

# calculo la anomalía de cada esfera por separado y luego las sumo por principio de superposición    

esf1 = funciones_utiles.anomZ_esf(rad,Dden,x1,y1,z1,x,y,zA)
esf2 = funciones_utiles.anomZ_esf(rad,Dden,x2,y2,z2,x,y,zA)

grav = (esf1 + esf2)
```

```python
#-----------------GRÁFICO--------------------
# ploteo la anomalía obtenida

cmap = "inferno"
plt.figure(1,figsize=(6,6))
plt.title("$Anomalía \ para \ z=0m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,grav*10**5,10,colors='grey')      # multiplico por 10**5 para pasar de [m/s2] a [mGal]
plt.imshow(grav*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.clim(np.min(grav*10**5), np.max(grav*10**5))
plt.show()
```

**Calculo la anomalía producida por dos esferas de radio 200 m para z=100**

Llamando a la función anom_esf definida en la note funciones_utiles

```python
zB     = 100.              # coordenada z donde calculamos la anomalía  [m]                                            

esf11  = funciones_utiles.anomZ_esf(rad,Dden,x1,y1,z1,x,y,zB)
esf21  = funciones_utiles.anomZ_esf(rad,Dden,x2,y2,z2,x,y,zB)

grav1  = (esf11 + esf21)
```

```python
#-----------------GRÁFICO--------------------
# ploteo la anomalía obtenida

cmap   = "inferno"
plt.figure(2,figsize=(6,6))
plt.title("$Anomalía \ para \ z=100m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,grav1*10**5,10,colors='grey')
plt.imshow(grav1*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.clim(np.min(grav*10**5), np.max(grav*10**5))
plt.show()
```

**Calculo la anomalía vertical producida por dos esferas de radio 200 m para z = - 20 m**

```python
zC= -20     #[m]

esf12 = funciones_utiles.anomZ_esf(rad,Dden,x1,y1,z1,x,y,zC)
esf22 = funciones_utiles.anomZ_esf(rad,Dden,x2,y2,z2,x,y,zC)

grav2 = (esf12 + esf22)

```

```python
#-----------------GRÁFICO--------------------
# ploteo la anomalía obtenida

cmap = "inferno"
plt.figure(1,figsize=(6,6))
plt.title("$Anomalía \ para \ z=-20m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,grav2*10**5,10,colors='grey')  #multiplico por 10**5 para pasar de [m/s2] a [mGal]
plt.imshow(grav2*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.clim(np.min(grav*10**5), np.max(grav*10**5))
plt.show()
```

```python
#---------------------------------GRÁFICO---------------------------------
# ploteo las 3 anomalías obtenidas juntas para poder comparar visualmente

cmap = "inferno"
fig, axs = plt.subplots(ncols=3) 

fig.set_figheight(20)
fig.set_figwidth(20)

extent = [-1100,1100,-1100,1100]

axs[0].set_title("$Anomalía \ para \ z=-20m$")
axs[0].set_xlabel("$x$")
axs[0].set_ylabel("$y$")
CS1 = axs[0].contour(x,y,grav2*10**5,10,colors='grey')  # multiplico por 10**5 para pasar de [m/s2] a [mGal]
im=axs[0].imshow(grav2*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')

axs[1].set_title("$Anomalía \ para \ z=0m$")
axs[1].set_xlabel("$x$")
axs[1].set_yticklabels([])                              # oculto los números del eje
CS2 = axs[1].contour(x,y,grav*10**5,10,colors='grey')   # multiplico por 10**5 para pasar de [m/s2] a [mGal]
axs[1].imshow(grav*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')

axs[2].set_title("$Anomalía \ para \ z=100m$")
axs[2].set_xlabel("$x$") 
axs[2].set_yticklabels([])                              # oculto los números del eje
CS3 = axs[2].contour(x,y,grav1*10**5,10,colors='grey')  # multiplico por 10**5 para pasar de [m/s2] a [mGal]
axs[2].imshow(grav1*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')

im.set_clim(np.min(grav*10**5), np.max(grav*10**5))
fig.colorbar(im, ax=axs, label="[mGal]", pad=0.1, aspect=50, orientation="vertical", shrink=0.3)  

# ax define a que grafica le asigno la barra, acá la defino para el conjunto y no a una en particular
# shrink modifica el tamaño de la barra
# aspect modifica la relacion ancho/largo de la barra
                
plt.show()
```

## Cálculo del campo anómalo por continuación analítica ascendente para $Dz=100m$ y su comparación con el campo para $z=100m$

**Calculo los números de ondas discretos de x e y. Además, calculo el número de onda radial para luego graficarlo.**

```python
kx = funciones_utiles.k_discreto(x)              #[1/m]
ky = funciones_utiles.k_discreto(y)              #[1/m]

nx = len(x)                                      # cantidad de muestras en x
ny = len(y)                                      # cantidad de muestras en y


# defino Kr antemano como ceros:

kr = np.zeros(nx*ny).reshape(nx,ny)


for i in range(nx):
    
        kr[i,:] = np.sqrt(kx[i]**2+ky[:]**2)  

        
```

```python
#-----------------GRÁFICO--------------------
# ploteo el número de onda radial discreto:

cmap="gray"

plt.figure(3,figsize=(6,6))
plt.title("$|k_r|$")
plt.xlabel("$k_x$")
plt.ylabel("$k_y$")
plt.xticks([0,np.pi,2*np.pi],[r'$0$',r'$\pi$',r'$2\pi$'])
plt.yticks([0,np.pi,2*np.pi],[r'$0$',r'$\pi$',r'$2\pi$'])
extent=[0,2*np.pi,0,2*np.pi]
plt.imshow(kr,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="Amplitud [1/m]", pad=0.1, aspect=100, orientation="vertical")
plt.show()
```

**Calculo la continuación ascendente para llevar la anomalía de z = 0 m a z = 100 m usando el número de onda radial Kr** 

```python

f    = grav.reshape(nx,ny)      # totalmente inecesario, grav ya la definimos con esa forma:  (grav==f).all() va a dar TRUE
Dz   = 100                      #[m] diferencia de altura, positivo porque estoy ascendiendo

#TDF 2D

F     = np.fft.fft2(f)

Fca   = np.exp(-Dz*kr)          # transformada del operador de continuidad ascendente
F_out = Fca*F

fca   = np.real(np.fft.ifft2(F_out))

```

```python
#-----------------------------GRÁFICO-----------------------------
# ploteo la trasnformada del operador de continuidad ascendente 

plt.figure(4,figsize=(6,6))
cmap="gray"
plt.title("$Transformada \ operador \ ascendente \ Dz=100m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.xticks([0,np.pi,2*np.pi],[r'$0$',r'$\pi$',r'$2\pi$'])
plt.yticks([0,np.pi,2*np.pi],[r'$0$',r'$\pi$',r'$2\pi$'])
extent=[0,2*np.pi,0,2*np.pi]
plt.imshow(Fca,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="Amplitud", pad=0.1, aspect=100, orientation="vertical")
plt.show()
```

```python
#-----------------GRÁFICO--------------------
# ploteo la continuación ascendente para un delta Z = 100 m 

plt.figure(5,figsize=(6,6))
plt.title("$Continuación \ ascendente  \ Dz=100m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
cmap="inferno"
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,fca*10**5,10,colors='grey')
plt.imshow(fca*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.clim(np.min(grav*10**5), np.max(grav*10**5))
plt.show()
```

```python
#-----------------GRÁFICO--------------------
# ploteo la continuación ascendente para delta Z=100m VS anomalía verdadera en Z=100m

fig, axs = plt.subplots(ncols=2) 

fig.set_figheight(15)
fig.set_figwidth(15)


axs[0].set_title("$Anomalía \ para \ z=100m$")
axs[0].set_xlabel("$x$") 
axs[0].set_ylabel("$y$")
CS1 = axs[0].contour(x,y,grav1*10**5,10,colors='grey')  # multiplico por 10**5 para pasar de [m/s2] a [mGal]
im=axs[0].imshow(grav1*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')

axs[1].set_title("$Continuación \ ascendente  \ Dz=100m$")
axs[1].set_xlabel("$x$") 
axs[1].set_yticklabels([])                              # oculto los números del eje
CS2 = axs[1].contour(x,y,fca*10**5,10,colors='grey')  # multiplico por 10**5 para pasar de [m/s2] a [mGal]
axs[1].imshow(fca*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')


cmap="inferno"
extent = [-1100,1100,-1100,1100]

im.set_clim(np.min(grav*10**5), np.max(grav*10**5))
fig.colorbar(im, ax=axs, label="[mGal]", pad=0.1, aspect=50, orientation="vertical", shrink=0.4)

plt.show()
```

**Ploteo la diferencia entre la anomalía verdadera de las esferas para z=100 m y la continuación ascendente con Dz= 100 m** 

```python
#-----------------GRÁFICO--------------------

plt.figure(5,figsize=(6,6))
plt.title("$[ \ Anom.\ verd. \ - \ Cont. \ asc.],  \ Dz=100m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
cmap="inferno"
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,(grav1-fca)*10**5,10,colors='grey')
plt.imshow((grav1-fca)*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.show()
```

## Cálculo del campo anómalo por continuación analítica descendente para $Dz=-20m$ y su comparación con el campo para $z=-20m$

**Calculo la continuación descendente para llevar la anomalía de z = 0 m a z = -20 m usando el número de onda radial Kr**

```python
f1    = grav.reshape(nx,ny)          # totalmente inecesario, grav ya la definimos con esa forma:  (grav==f).all() va a dar TRUE
Dz1   = -20                          #[m] diferencia de altura, negativo porque estoy descendiendo

#TDF 2D

F1 = np.fft.fft2(f1)

Fca1 = np.exp(-Dz1*kr)               # transformada del operador de continuidad descendente
F_out1 = Fca1*F1              

fca1=np.real(np.fft.ifft2(F_out1))

```

```python
#-----------------GRÁFICO--------------------
# ploteo la trasnformada del operador de continuidad descendente 

plt.figure(4,figsize=(6,6))
cmap="gray"
plt.title("$Transformada \ operador \ descendente \ Dz=-20m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.xticks([0,np.pi,2*np.pi],[r'$0$',r'$\pi$',r'$2\pi$'])
plt.yticks([0,np.pi,2*np.pi],[r'$0$',r'$\pi$',r'$2\pi$'])
extent=[0,2*np.pi,0,2*np.pi]
plt.imshow(Fca1,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="Amplitud", pad=0.1, aspect=100, orientation="vertical")
plt.show()
```

```python
#-----------------GRÁFICO--------------------
# ploteo la continuación descendente para un delta Z = -20 m 

plt.figure(5,figsize=(6,6))
plt.title("$Continuación \ descendente \ Dz=-20m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
cmap="inferno"
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,fca1*10**5,10,colors='grey')
plt.imshow(fca1*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.clim(np.min(grav*10**5), np.max(grav*10**5))
plt.show()
```

```python
#-----------------GRÁFICO--------------------
# ploteo la continuación descendente para delta Z=-20m VS anomalía verdadera en Z=-20m

fig, axs = plt.subplots(ncols=2) 

fig.set_figheight(15)
fig.set_figwidth(15)


axs[0].set_title("$Anomalía \ para \ z=-20m$")
axs[0].set_xlabel("$x$") 
axs[0].set_ylabel("$y$")
CS1 = axs[0].contour(x,y,grav2*10**5,10,colors='grey')  # multiplico por 10**5 para pasar de [m/s2] a [mGal]
im=axs[0].imshow(grav2*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')

axs[1].set_title("$Continuación \ descendente  \ Dz=-20m$")
axs[1].set_xlabel("$x$") 
axs[1].set_yticklabels([])                              # oculto los números del eje
CS2 = axs[1].contour(x,y,fca1*10**5,10,colors='grey')  # multiplico por 10**5 para pasar de [m/s2] a [mGal]
axs[1].imshow(fca1*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')


cmap="inferno"
extent = [-1100,1100,-1100,1100]

im.set_clim(np.min(grav*10**5), np.max(grav*10**5))
fig.colorbar(im, ax=axs, label="[mGal]", pad=0.1, aspect=50, orientation="vertical", shrink=0.4)

plt.show()
```

**Ploteo la diferencia entre la anomalía verdadera de las esferas para z=-20 m y la continuación descendente con Dz= -20 m**

```python
#-----------------GRÁFICO--------------------

plt.figure(5,figsize=(6,6))
plt.title("$[ \ Anom.\ verd. \ - \ Cont. \ desc.],  \ Dz=-20m$")
plt.xlabel("$x$")
plt.ylabel("$y$")
cmap="inferno"
extent = [-1100,1100,-1100,1100]
CS = plt.contour(x,y,(grav2-fca1)*10**5,5,colors='gainsboro')
plt.imshow((grav2-fca1)*10**5,origin="lower",cmap=cmap,extent=extent,interpolation='none')
plt.colorbar(label="[mGal]", pad=0.1, aspect=100, orientation="vertical")
plt.show()
```
